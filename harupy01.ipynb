```python
def multis(x2, y2):
    resting = []
    res1 = x2 + y2
    res2 = x2 - y2
    res3 = x2 * y2
    res4 = x2 // y2
    res5 = x2 % y2
    res6 = x2 ** y2
    resting.append(res1)
    resting.append(res2)
    resting.append(res3)
    resting.append(res4)
    resting.append(res5)
    resting.append(res6)
    return resting

Lists = []
ekk1, ekk2 = 0, 0

Lists = multis(100, 20)

print("이것이 코어다.",(Lists))
```

```python
lists = [67, 69, 74]
print("현재 리스트: %s"%lists)

lists.append(999)
print("append 후 리스트: %s"%lists)

print("pop으로 추출한 값: %s"%lists.pop())
print("pop 후 리스트: %s"%lists)

lists.sort()
print("sort 후 리스트: %s"%lists)

lists.reverse()
print("reverse 후 리스트: %s"%lists)

lists.insert(1, 222)
print("insert 후 리스트: %s"%lists)

lists.remove(222)
print("remove 후 리스트: %s"%lists)

lists.extend([12, 34, 67])
print("extend 후 리스트: %s"%lists)

print("67값의 개수: %d"%lists.count(67))
```

```python
list1 = []
list2 = []
value = 1
for i in range(0,10):
    for k in range (0,10):
        list1.append(value)
        value +=1
    list2.append(list1)
    list1 = []

for i in range(0,10):
    for k in range (0,10):
        print("%3d" % list2[i][k], end= " ")
    print(" ")
```

```python
list3 = []
list4 = []
value = 100
for i in range(0,10):
    for k in range (0,10):
        list3.append(value)
        value -=1
    list4.append(list3)
    list3 = []

for i in range(0,10):
    for k in range (0,10):
        print("%3d" % list4[i][k], end= " ")
    print(" ")

total_sum = 0
for row in list4:
    for num in row:
        total_sum += num

print(f"배열의 합계 ==> {total_sum}")
```
