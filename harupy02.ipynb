```python
def priPol(ppx):
    term = len(ppx) - 1 
    Str = "P(x) = "

    for i in range(len(ppx)):
        coef = ppx[i]

        if coef == 0:
            term -= 1
            continue

        if coef >= 0 and i > 0: 
            Str += " + "
        elif coef < 0:
            Str += " " 
            
        Str += str(coef) + "x^" + str(term)
        term -= 1
    return Str

def calPol(xVal, ppx):
    resultVal = 0 
    term = len(ppx) - 1
    
    for i in range(len(ppx)): 
        coef = ppx[i]
        resultVal += coef * (xVal ** term)
        term -= 1
    return resultVal

px = [6, -9, 0, 4]

if __name__ == "__main__":
    pStr = priPol(px)
    print(pStr)

    xValue = int(input("x 값: "))

    pxValue = calPol(xValue, px)
    print(f"결과: {pxValue}")
```
